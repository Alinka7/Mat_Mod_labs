```python
from pylab import *
from scipy.integrate import *
t = linspace(0,40*pi,10000)
x = [0.0, 0.0]
w = 0.3
lamda = 1
eps = 0.1
def mateu(y,t):
    x,v = y
    return [v,-x*(w*w+eps*cos(lamda*t))]

def plot_f(func):
    result = odeint(mateu,[1.0,1.0],t)
    x = result[:,0]
    v = result[:,1]
    plot(t,x,lw=0.8)
    xlabel('t')
    ylabel('x')
    grid()

def port (func):
    result = odeint(mateu,[1.0,1.0],t)
    plot(result[:,0], result[:,1], lw=0.8)
    xlabel('x')
    ylabel('v')
    grid()
```

```python
port(mateu)
```

```python
plot_f(mateu)
```

```python
from pylab import *
from scipy.integrate import *
t = linspace(0,40*pi,10000)
x = [1.0, 1.0]
w = 1
m = 1
k = 1
lamda = 1
e = 0.0
def vanDerPole(y,t):
    x,v = y
    return [y[1],-w*w*y[0]+m*(1-y[0]*y[0])*y[1]+k*y[0]*cos(lamda*t)]

def plot_f(func):
    result = odeint(vanDerPole,[1.0, 1.0],t)
    x = result[:,0]
    v = result[:,1]
    plot(t,x,lw=0.8)
    xlabel('t')
    ylabel('x')
    grid()
    
def port (func):
    result = odeint(vanDerPole,[1.0,1.0],t)
    plot(result[:,0], result[:,1], linewidth=0.6)
    xlabel('x')
    ylabel('v')
    grid()
```
