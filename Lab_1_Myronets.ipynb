#  Фігури Лісажу 

```python
from numpy import pi,sin
def Lissajous_curve():
    w = [[0],[4,4],[7,4],[3,6],[9,5]]
    for i in w:
        if i[0]==0:
            a=8
            x=[sin(a*t) for t in np.arange(0.,2*pi,0.01)]
            y=[sin(a*t) for t in np.arange(0.,2*pi,0.01)]
            plt.plot(x, y, 'g')
            plt.grid (True)
            plt.show()
        else:
            w = [[4,4],[7,4],[3,6],[9,5]]
            for i in w:
                a = i[0]
                b = i[1]
                f=0.5*pi
                x=[sin(a*t+f) for t in np.arange(0.,2*pi,0.01)]
                y=[sin(b*t) for t in np.arange(0.,2*pi,0.01)]
                plt.plot(x, y, 'g')
                plt.grid (True)
                plt.show()
Lissajous_curve()
```


# Поліном Чебишева

```python
def polynom(x, n):
    return np.cos(n * np.arccos(x))
def chebyshev():
    x = np.linspace(-1, 1, 1000)
    plt.figure(figsize=(15,7))
    #plt.subplot(2,3,5)
    plt.plot(x, polynom(x, 6),'r')
chebyshev()    
```

# Годограф Михайлова

```python
def polynom(p):    
    return 17*p**5 + 4*p**3 + 5*p**2 + 21*p + 13

def mikhailov(polynom, w):
    wj = 1j * w
    d = polynom(wj)
    return d.real, d.imag

w = np.linspace(0, 1000, 10000)
x, y = mikhailov(polynom, w)

def plot_mikhailov(x, y, xlim, ylim):
    plt.xlim(xlim[0], xlim[1])
    plt.ylim(ylim[0], ylim[1])
    plt.plot(x,y)
    plt.grid()

plt.figure(figsize=(15,10))

plt.subplot(1,3,3)
plot_mikhailov(x, y, (0, 30), (0,250))
```
